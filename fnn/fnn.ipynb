```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
def relu(x):
    return x * (x > 0)
```

```python
N1, N2 = 4, 5
epochs, sigma = 10, .05
```

```python
# load data
X, y = datasets.load_boston(return_X_y=True)
[n, dim] = X.shape

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25)
```

```python
[n_train, dim] = X_train.shape
```

```python
n, dim
```

```python
# init weights
W1 = np.random.rand(dim + 1, N1) - .5
W2 = np.random.rand(N1 + 1, N2) - .5
w = np.random.rand(N2 + 1) - .5
```

```python
for i in range(epochs):
    A1 = relu(np.matmul(np.column_stack([X_train, np.ones(n_train)]), W1))
    A2 = relu(np.matmul(np.column_stack([A1, np.ones(n_train)]), W2))
    y_hat = np.matmul(np.column_stack([A2, np.ones(n_train)]), w)
    loss = np.dot(y_hat - y_train, y_hat - y_train)
    print('Epoch = ' + str(i) + ', loss = ' + str(loss))
    
    # update weights
    w -= sigma * 2 * np.matmul((y_hat - y_train), np.column_stack([A2, np.ones(n_train)]))
    
```
